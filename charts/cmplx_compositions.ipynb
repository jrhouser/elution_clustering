```python
%matplotlib inline

import pandas as pd
import numpy as np
import pylab as plt
import cPickle

cmplxs_GMM_bayes=cPickle.load(open('../testdata/PA/GMM_bayes_0.9_cluster_filterones.p','rb'))
cmplxs_GMM=cPickle.load(open('../testdata/PA/GMM_cluster_compare_ms1_PA_clusters_filterones.p','rb'))[0]

print 'number of gmm bayes clusters with more than one protein =' + str(len(cmplxs_GMM_bayes))
print 'number of gmm clusters with more than one protein =' + str(len(cmplxs_GMM))


punique_bayes=len(set([p for c in cmplxs_GMM_bayes for p in c]))
punique_gmm=len(set([p for c in cmplxs_GMM for p in c]))

print 'number of unique proteins in gmm bayes clusters = ' + str(punique_bayes)
print 'number of unique proteins in gmm clusters = ' + str(punique_gmm)

l_bayes=[len(c) for c in cmplxs_GMM_bayes]
l_gmm=[len(c) for c in cmplxs_GMM]
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
plt.hist(l_bayes,bins=100)
plt.title('Bayes GMM',fontsize=18)
plt.xlim([0,150])
plt.ylim([0,85])
plt.subplot(2,1,2)
plt.hist(l_gmm,bins=300)
plt.title('GMM',fontsize=18)
plt.xlabel('complex size',fontsize=18)
plt.xlim([0,150])
plt.ylim([0,85])
plt.show()
```
