```python
%matplotlib inline
import pylab as plt
import pandas as pd
import numpy as np
import cPickle
import sys
sys.path.append('../')
import GMM

#load data

df_elutions=GMM.concatenate_elutions('../testdata/PA/ms1_filtered/')

#import complexes 
cmplxs_GMM_bayes=cPickle.load(open('../testdata/PA/GMM_bayes_0.9_cluster_filterones.p','rb'))
cmplxs_GMM=cPickle.load(open('../testdata/PA/GMM_cluster_compare_ms1_PA_clusters_filterones.p','rb'))[0]

#import gene names
df_annotate=pd.read_table('../testdata/PA/PA14_Uniprot_and_go_acc_improved.tab').set_index('Unnamed: 0')
```

```python
def get_cmplx(cmplxs,cmplx_id='random'):
    
    if cmplx_id=='random':
        cmplx_id=np.random.randint(0,len(cmplxs))
    df=df_elutions.ix[cmplxs[cmplx_id]]
    
    return df
```

```python
df=get_cmplx(cmplxs_GMM,cmplx_id='random') #cmplx_id can be an interger value


x=np.arange(len(df.columns))
plt.figure(figsize=(20,20))
for i,ix in enumerate(df.index):
    d=df.ix[ix]
    
    if ix in df_annotate.index:
        gn=df_annotate.loc[ix,'Gene Name']
    else:
        gn=ix
        
    ax= plt.subplot(len(df),1,i+1)
    plt.ylabel(gn,fontsize=18,rotation='horizontal')
    ax.yaxis.labelpad = 100
    plt.bar(x,d)
    plt.xlim([0,max(x)])
    plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off

    
    
plt.show()
```
